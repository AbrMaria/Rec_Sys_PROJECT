```python
import json
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
with open('coursera_coursera_reviews.json') as data_file:
    coursera_reviews_df = json.load(data_file)
```

```python
coursera_reviews_df
```

```python
coursera_reviews_df = DataFrame(coursera_reviews_df)
```

```python
coursera_reviews_df = pd.DataFrame(coursera_reviews_df, columns= ['course_id',
                                                  'reviewer_name',
                                                  'rating',
                                                  'review_text',
                                                  'url',])
```

```python
coursera_reviews_df.info()
```

```python
coursera_reviews_df['course_id'] = coursera_reviews_df['course_id'].astype(str)
coursera_reviews_df['reviewer_name'] = coursera_reviews_df['reviewer_name'].astype(str)
coursera_reviews_df['review_text'] = coursera_reviews_df['review_text'].astype(str)
coursera_reviews_df['url'] = coursera_reviews_df['url'].astype(str)
```

```python
coursera_reviews_df.to_csv('coursera_users_reviews_df.csv')
```
